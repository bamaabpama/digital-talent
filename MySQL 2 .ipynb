### One to many relationship
1. Report the account representative for each customer.
2. Report total payments for Atelier graphique.
3. Report the total payments by date
4. Report the products that have not been sold.
5. List the amount paid by each customer.
6. How many orders have been placed by Herkku Gifts?
7. Who are the employees in Boston?
8. Report those payments greater than \$100,000. Sort the report so the customer who made the highest payment appears first.
9. List the value of 'On Hold' orders.
10. Report the number of orders 'On Hold' for each customer.


```python
import pymysql
import pandas as pd

host = '127.0.0.1'
port = "3306"
user = 'root'
password = ''
database = 'classicmodels'

conn = pymysql.connect(
    host=host,
    port=int(port),
    user=user,
    passwd=password,
    db=database,
    charset='utf8mb4')

def run(sql):
    df = pd.read_sql_query(sql,conn)
    return df
```

```python
# 1. Report the account representative for each customer.

run("""
SELECT
    CONCAT(firstName,' ',lastName) AS fullName,
    e.employeeNumber,
    c.customerNumber
FROM
    employees e 
        INNER JOIN 
        customers c ON e.employeeNumber = c.SalesRepEmployeeNumber
""")
```

```python
# 2. Report total payments for Atelier graphique.
# run("SELECT customerNumber,customerName FROM customers WHERE customerName='Atelier graphique'")
run("""
SELECT
    p.customerNumber,
    customerName,
    SUM(amount) AS totalPayments
FROM
    payments p
        INNER JOIN
    customers c ON p.customerNumber=c.customerNumber
WHERE
    c.customerName='Atelier graphique'
""")
```

```python
# 3. Report the total payments by date
run("""
SELECT SUM(amount),paymentDate FROM payments GROUP BY paymentDate
""")
```

```python
# 4. Report the products that have not been sold.
run("""
SELECT
    productCode,
    productName
FROM products P
""")
```

```python
# 5. List the amount paid by each customer.
run("""
SELECT
    p.customerNumber,
    customerName,
    SUM(amount)
FROM
    payments p
        LEFT JOIN
    customers c ON p.customerNumber=c.customerNumber
GROUP BY customerNumber
""")
```

```python
# 6. How many orders have been placed by Herkku Gifts?
run("""
SELECT
    o.customerNumber,
    customerName,
    COUNT(status) AS orderShipped
FROM
    orders o
        INNER JOIN
    customers c ON o.customerNumber=c.customerNumber
WHERE
    customerName='Herkku Gifts'
""")
```

```python
# 7. Who are the employees in Boston?
run("""
SELECT
    employeeNumber,
    CONCAT (firstName,' ',lastName) As employeeName
FROM
    employees e
        INNER JOIN
    offices o ON e.officeCode=o.officeCode
WHERE city ='Boston'
""")
```

```python
# 8. Report those payments greater than \$100,000. Sort the report so the customer who made the highest payment appears first.
run("""
SELECT
    c.customerNumber,
    c.customerName,
    amount
FROM
    customers c
        INNER JOIN
    payments p ON c.customerNumber=p.customerNumber
WHERE
    amount>100000
ORDER BY amount DESC
""")
```

```python
# 9. List the value of 'On Hold' orders.
run("""
SELECT od.orderNumber,(quantityOrdered*priceEach) AS value
FROM orderdetails od
INNER JOIN orders o
ON od.orderNumber=o.orderNumber
WHERE status='On Hold'
""")
```

```python
# 10. Report the number of orders 'On Hold' for each customer.
run("""
SELECT o.customerNumber,customerName, count(status) AS orderStatus
FROM customers c
INNER JOIN orders o
ON c.customerNumber=o.customerNumber
WHERE status='On Hold'
GROUP BY o.customerNumber
""")
```

### Many to many relationship
1. List products sold by order date.
2. List the order dates in descending order for orders for the 1940 Ford Pickup Truck.
3. List the names of customers and their corresponding order number where a particular order from that customer has a value greater than $25,000?
4. Are there any products that appear on all orders?
5. List the names of products sold at less than 80% of the MSRP.
6. Reports those products that have been sold with a markup of 100% or more (i.e.,  the priceEach is at least twice the buyPrice)
7. List the products ordered on a Monday.
8. What is the quantity on hand for products listed on 'On Hold' orders?

```python
# 1. List products sold by order date.
run("""
SELECT
    orderDate,
    productName
FROM
    products 
        INNER JOIN orderdetails USING (productCode)
        INNER JOIN orders USING (orderNumber)
ORDER BY orderDate
""")
```

```python
# 2. List the order dates in descending order for orders for the 1940 Ford Pickup Truck.
run("""
SELECT
    orderDate,
    productName
FROM
    products 
        INNER JOIN orderdetails USING (productCode)
        INNER JOIN orders USING (orderNumber)
WHERE productName='1940 Ford Pickup Truck'
ORDER BY orderDate
""")
```

```python
# 3. List the names of customers and their corresponding order number where a particular order from that customer has a value greater than $25,000?
run("""
SELECT customerName,orderNumber,SUM(quantityOrdered*priceEach) As value
FROM customers
    INNER JOIN orders USING(customerNumber)
    INNER JOIN orderDEtails USING(orderNumber)
GROUP BY orderNumber
HAVING value>25000
""")
```

```python
# 4. Are there any products that appear on all orders?
run("""
SELECT  productCode,COUNT(*) AS banyak
FROM orderDetails
GROUP BY productCode
HaVING banyak = (SELECT COUNT(*) FROM orders)
""")
```

```python
# 5. List the names of products sold at less than 80% of the MSRP.
run("""
SELECT orderNumber,productName,priceEach,MSRP
FROM products
INNER JOIN orderdetails USING(productCode)
WHERE priceEach >= (0.8*MSRP)
""")
```

```python
# 6. Reports those products that have been sold with a markup of 100% or more (i.e.,  the priceEach is at least twice the buyPrice)
run("""
SELECT productCode,productName,priceEach,buyPrice,(buyPrice*2) AS markup
FROM products
INNER JOIN orderdetails USING (productCode)
HAVING priceEach > markup
""")
```

```python
# 7. List the products ordered on a Monday.
run("""
SELECT DAYNAME(orderDate),productName
FROM products
INNER JOIN orderdetails USING(productCode)
INNER JOIN orders USING (orderNumber)
WHERE DAYNAME(orderDate)='Monday'
""")
```

```python
# 8. What is the quantity on hand for products listed on 'On Hold' orders?
run("""
SELECT SUM(quantityOrdered) AS Quantity
FROM orders
INNER JOIN orderdetails USING(orderNumber)
WHERE status='On Hold'
""")
```
