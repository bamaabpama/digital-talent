```python
import pymysql
import pandas as pd

host = '127.0.0.1'
port = "3306"
user = 'root'
password = ''
database = 'classicmodels'

conn = pymysql.connect(
    host=host,
    port=int(port),
    user=user,
    passwd=password,
    db=database,
    charset='utf8mb4')

def run(sql):
    df = pd.read_sql_query(sql,conn)
    return df
```

```python
# Prepare a list of offices sorted by country, state, city.
run("SELECT country,state,city FROM offices")
```

```python
# How many employees are there in the company?
run("SELECT COUNT(employeeNumber) AS jumlahKaryawan FROM employees")
```

```python
# What is the total of payments received?

run("SELECT SUM(amount) AS totalPayments FROM payments")
```

```python
# List the product lines that contain 'Cars'.
# run("SELECT productLine FROM productlines")
run("SELECT productLine FROM productlines WHERE productLine LIKE '%Cars%'")
```

```python
# Report total payments for October 28, 2004.
run("SELECT SUM(amount) AS totalPayments FROM payments WHERE paymentDate='2004-10-28'")
```

```python
# Report those payments greater than $100,000.


run("SELECT customerNumber,amount FROM payments WHERE amount>100000")
```

```python
# List the products in each product line.

run("SELECT productCode,productName,productLine FROM products GROUP BY productLine")
```

```python
# How many products in each product line?

run("SELECT productLine,COUNT(productLine) AS jumlahProductLine FROM products GROUP BY productLine")
```

```python
# What is the minimum payment received?
run("SELECT customerNumber,checkNumber,paymentDate,MIN(amount) AS minimumPayment FROM payments")
```

```python
# List all payments greater than twice the average payment.
run("SELECT customerNumber,checkNumber,paymentDate,amount FROM payments WHERE amount>(SELECT AVG(amount)*2 AS 2TotalPayment FROM payments)")
```

```python
# What is the average percentage markup of the MSRP on buyPrice?
# Gross Profit Margin = Sales Price – Unit Cost = $125 – $100 = $25
# Markup Percentage = Gross Profit Margin/Unit Cost = $25/$100 = 25%
run("SELECT AVG((MSRP-buyPrice)/buyPrice*100) AS markupPercentage FROM products")
```

```python
# How many distinct products does ClassicModels sell?
# run("SELECT * FROM products")
run("SELECT COUNT(DISTINCT productLine) AS jumlahDistinctProduct FROM products")
```

```python
# Report the name and city of customers who don't have sales representatives?
run("SELECT customerName,city FROM customers WHERE salesRepEmployeeNumber IS NULL")
# run("SELECT customerName,city,salesRepEmployeeNumber FROM customers")
```

```python
# What are the names of executives with VP or Manager in their title? Use the CONCAT function to combine the employee's first name and last name into a single field for reporting.
run("SELECT CONCAT(firstName,' ',lastName) AS fullName,jobTitle FROM employees WHERE jobTitle LIKE '%VP%' OR jobTitle LIKE '%Manager%'")
```
